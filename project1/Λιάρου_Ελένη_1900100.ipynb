 Επομένως έχουμε $\nabla_f(x) = \nabla_x(x^TAx+b^Tx) = \frac{\partial x^TAx}{\partial x} + \frac{\partial b^Tx}{\partial x} = (Α+Α^Τ)x + b = 2Ax+b$
* 2.2 \
  έχουμε $f(x) = ||A-XB||_F^2 = Tr((A-XB)(A-XB)^T) = Tr((A-XB)(A^T-(XB)^T)) = Tr((A-XB)(A^T-B^TX^T))= Tr((A-XB)A^T-(A-XB)B^TX^T) = Tr(AA^T-XBA^T-AB^TX^T+XBB^TX^T)=Tr(AA^T)-Tr(XBA^T)-Tr(AB^TX^T)+Tr(XBB^TX^T)$ 

 η παράγωγος της f είναι: $\frac{\partial f(x)}{\partial x} = 0-(BA^T)^T-AB^T+X(BB^T)^T+XBB^T = -AB^T-AB^T+XBB^T+XBB^T= -2AB^T+2XBB^T = -2(A-XB)B^T$ \
  Η συνάρτηση είναι κυρτή άρα το σημείο που μηδενίζεται η παράγωγός της θα είναι και ολικό ελάχιστο της f. Άρα το ολικό ελάχιστο της f είναι και η βέλτιστη λύση του προβλήματος της άσκησης. \

3. Ερώτημα 3 \
Ακολουθεί σε python η εφαρμογή του αλγορίθμου gradient descent στις συναρτήσεις της εκφώνησης. 

```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x1,x2):
  return (x1-2)**2 + (x2-3)**2

def f2(x1, x2):
  a = (1-(x2-3))**2
  b = 20*((x1+3)-(x2-3)**2)**2
  return a+b

def df1dx1(x1,x2):
  return 2*(x1-2)

def df1dx2(x1,x2):
  return 2*(x2-3)

def df2dx2(x1,x2):
  d = (x2-3)*(x1-(x2-3)**2 + 3)
  a = (-40)*d
  b = x2-4
  c = a + b
  return 2*c

def df2dx1(x1, x2):
  a = x1-(x2-3)**2 + 3
  return 40*a


#prota gia thn f1
print("showing results of gradient descent algorithm on f1 function")
curX1 = 0.0 # The algorithm starts at x=0
curX2 = 0.0
rate = 0.5 # Learning rate
precision = 0.0001 #This tells us when to stop the algorithm
previous_step_size = 1 #
maxIters = 10 # maximum number of iterations
iters = 0 #iteration counter
plt.figure()
for aIter in range(maxIters):
  prevX1=curX1
  curX1=curX1-rate*df1dx1(curX1, curX2)
  diff1=abs(curX1-prevX1)
  prevX2=curX2
  curX2=curX2-rate*df1dx2(curX1, curX2)
  diff2=abs(curX2-prevX2)
  if diff1<=precision and diff2<=precision:
    print("The local minimum occurs at (",curX1,",",curX2,")")
    break
  plt.scatter(aIter, f1(curX1, curX2))
  print("Iteration",aIter,"\n(X1,X2) value is (",curX1,",",curX2,")", "f1(curX1, curX2)=",f1(curX1, curX2)) #Print iterations

plt.xlabel('x - iterations')
plt.ylabel('y - f1(x1,x2)')
plt.title('Gradient descent of f1')
plt.show()


print("\n\n\n")


#meta gia thn f2
print("showing results of gradient descent algorithm on f2 function")
curX1 = 0.0 # The algorithm starts at x=0
curX2 = 0.0
rate = 0.5 # Learning rate
precision = 0.0001 #This tells us when to stop the algorithm
previous_step_size = 1 #
maxIters = 100000 # maximum number of iterations
iters = 0 #iteration counter
plt.figure()
a = []
b = []
for aIter in range(maxIters):
  prevX1=curX1
  curX1=curX1-rate*df2dx1(curX1, curX2)
  diff1=abs(curX1-prevX1)
  prevX2=curX2
  curX2=curX2-rate*df2dx2(curX1, curX2)
  diff2=abs(curX2-prevX2)
  a.append(aIter)
  b.append(f2(curX1, curX2))
  if diff1<=precision and diff2<=precision:
    print("The local minimum occurs at (",curX1,",",curX2,")")
    break
  print("Iteration",aIter,"\n(X1,X2) value is (",curX1,",",curX2,")", "f2(curX1, curX2)=",f2(curX1, curX2)) #Print iterations

plt.xlabel('x - iterations')
plt.ylabel('y - f2(x1,x2)')
plt.title('Gradient descent of f2')
plt.plot(a, b)
plt.show()
  
  


```
